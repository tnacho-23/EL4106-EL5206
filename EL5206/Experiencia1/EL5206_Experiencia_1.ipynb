```python
#Importacion de librerias
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Pregunta 1

```python
#Cargar y mostrar imagen "tomates.png"
tomates_bgr = cv2.imread("01_2024/tomates.png")
tomates = cv2.cvtColor(tomates_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(tomates)
```

```python
#Transforma imagen a escala de grises
tomates_gray = cv2.cvtColor(tomates,cv2.COLOR_RGB2GRAY)

plt.imshow(tomates_gray, cmap="gray")
```

```python
#Calculo histograma imagen gris
hist = cv2.calcHist([tomates_gray], [0], None, [256], [0, 256])
plt.plot(hist, color='gray' )
plt.title('Histograma tomates')
plt.xlabel('Intensidad')
plt.ylabel('Frecuencia')
plt.show()
```

```python
#Obtención de imagen binaria dependiente de un threshold "t"
# Define el umbral (t)
t = 125

# Aplica el thresholding
_, tomates_binary = cv2.threshold(tomates_gray, t, 255, cv2.THRESH_BINARY)

# Muestra la imagen binarias
plt.imshow(tomates_binary, cmap='gray')
```

```python
#Obtención de imagen binaria dependiente de un threshold "t"
# Define el umbral (t)
t = 100

# Aplica el thresholding
_, tomates_binary = cv2.threshold(tomates_gray, t, 255, cv2.THRESH_BINARY)

# Muestra la imagen binarias
plt.imshow(tomates_binary, cmap='gray')
```

```python
#Obtención de imagen binaria dependiente de un threshold "t"
# Define el umbral (t)
t = 25

# Aplica el thresholding
_, tomates_binary = cv2.threshold(tomates_gray, t, 255, cv2.THRESH_BINARY)

# Muestra la imagen binarias
plt.imshow(tomates_binary, cmap='gray')
```

```python
#Obtener canales BGR
r = tomates[:,:,0]
g = tomates[:,:,1]
b = tomates[:,:,2]
# Visualize with color mappings
f, axarr = plt.subplots(1, 3)
axarr[0].imshow(r, cmap="Reds")
axarr[1].imshow(g, cmap="Greens")
axarr[2].imshow(b, cmap="Blues")
plt.show()


```

```python
plt.imshow(r, cmap='gray')
plt.title('Canal rojo')
```

```python
plt.imshow(g, cmap='gray')
plt.title('Canal verde')
```

```python
plt.imshow(b, cmap='gray')
plt.title('Canal azul')
```

```python
fixed_aji = cv2.subtract(r, g)
plt.imshow(fixed_aji, cmap='gray')
```

```python
#Calculo histograma ají
hist = cv2.calcHist([fixed_aji], [0], None, [256], [0, 256])
plt.plot(hist, color='gray' )
plt.title('Histograma ají')
plt.xlabel('Intensidad')
plt.ylabel('Frecuencia')
plt.show()
```

```python
t = 90
_, mask = cv2.threshold(fixed_aji, t, 255, cv2.THRESH_BINARY)
# Muestra la imagen binarias
plt.imshow(mask, cmap='gray')
```

```python
img_out = cv2.bitwise_and(tomates, tomates, mask=mask)
plt.imshow(img_out)
```

# Pregunta 3

```python
#Cargar y mostrar imagen "fence.png"
fence_bgr = cv2.imread("01_2024/fence.png")
fence = cv2.cvtColor(fence_bgr, cv2.COLOR_BGR2RGB)
fence_gray = cv2.cvtColor(fence, cv2.COLOR_RGB2GRAY)

plt.imshow(fence_gray, cmap='gray')
```

```python
fence_fft = np.fft.fft2(fence_gray)
fence_fft_magnitude = np.abs(fence_fft)

plt.imshow(fence_fft_magnitude, cmap='gray')
plt.title('Magnitud de la Transformada de Fourier')
plt.show()
```

```python
R = np.max(fence_gray)
print(R)
c = 255 / np.log(1 + R)
print(c)
transf_log = c * np.log(1 + fence_fft)
'''
Como se vio en clases, fft shift hace que las frecuencias bajas esten en el centro
y las altas en las esquinas, para mejor interpretacion
'''
shift = np.fft.fftshift(transf_log)
abs_transf_log = np.abs(shift)

plt.imshow(abs_transf_log, cmap='gray')
plt.title('Magnitud de la Transformada de Fourier (con transformación logarítmica)')
plt.show()
```

```python
#Cargar y mostrar imagen "diagonal.png"
diagonal_bgr = cv2.imread("01_2024/diagonal.jpg")
diagonal = cv2.cvtColor(diagonal_bgr, cv2.COLOR_BGR2RGB)
diagonal_gray = cv2.cvtColor(diagonal, cv2.COLOR_RGB2GRAY)

plt.imshow(diagonal_gray, cmap='gray')
```

```python
diagonal_fft = np.fft.fft2(diagonal_gray)
diagonal_fft_magnitude = np.abs(diagonal_fft)

plt.imshow(diagonal_fft_magnitude, cmap='gray')
plt.title('Magnitud de la Transformada de Fourier')
plt.show()
```

```python
R = np.max(diagonal_gray)
print(R)
c = 255 / np.log(1 + R)
print(c)
transf_log = c * np.log(1 + diagonal_fft)
shift = np.fft.fftshift(transf_log)
abs_transf_log = np.abs(shift)

plt.imshow(abs_transf_log, cmap='gray')
plt.title('Magnitud de la Transformada de Fourier (con transformación logarítmica)')
plt.show()
```
